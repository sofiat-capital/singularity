```python
api = MasterAPI.MasterAPI()
```

```python
#api.DataBaseAPI.DayCandle

DayCandle = api.DataBaseAPI.DayCandle
session = api.DataBaseAPI.engine.Session()

candles = session.query(DayCandle).filter(DayCandle.fk_idproduct_dayCandle == 4).all()
```

```python
reformat = []
for candle in candles:
    
    reformat.append([candle.date, candle.close, candle.volume])
```

```python
frame = pd.DataFrame(reformat, columns = ['date', 'close', 'volume'])
```

```python
#frame.iloc[:,1].rolling(window=3).mean()

np.array(frame.iloc[:,1].rolling(window=3).mean()).shape
```

```python
palette = cm.viridis
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))

ax.plot(frame['date'], frame['close'], color = 'black')
ax.plot(frame['date'], np.array(frame.iloc[:,1].rolling(window=30).mean()), label = '30 days')
ax.plot(frame['date'], np.array(frame.iloc[:,1].rolling(window=200).mean()), label = '200 days')


ax.set_ylabel("Price [$]")
ax.set_xlabel('Date')

ax.legend()

plt.savefig('ETH_MACD.png')
plt.show()
```
